```python
import pandas as pd
```

```python
kg_data = pd.read_csv('/data/zhaojingtong/PrimeKG/data_all/kg_directed.csv')
kg_data
```

```python
df_x_data = kg_data[['x_type','x_name','x_idx']].rename(columns = {'x_type':'type','x_name':'name','x_idx':'idx'})
df_x_data
```

```python
df_y_data = kg_data[['y_type','y_name','y_idx']].rename(columns={'y_type':'type','y_name':'name','y_idx':'idx'})
df_y_data
```

```python
all_data = pd.concat([df_x_data,df_y_data])
all_data
```

```python
all_data_uni = all_data.drop_duplicates().reset_index(drop=True)
all_data_uni['idx'] = all_data_uni['idx'].astype(int)
all_data_uni
```

```python
drug = all_data_uni[all_data_uni['type'] == 'drug'].reset_index(drop=True)
drug
```

```python
drug[drug['idx'] == 7957]
```

```python
disease = all_data_uni[all_data_uni['type'] == 'disease'].reset_index(drop=True)
disease
```

```python
from transformers import BertModel, BertTokenizer
import pickle
import torch
from tqdm import tqdm

# 检查CUDA是否可用，选择设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 加载BioBERT模型和Tokenizer
model_path = '/data/zhaojingtong/Biobert_embedding/biobert'
tokenizer = BertTokenizer.from_pretrained(model_path, do_lower_case=False)
model = BertModel.from_pretrained(model_path)

# 将模型移到GPU（如果可用）
model = model.to(device)

embeddings_dict = {}

# 为每个drug的名称生成embedding
for _, row in tqdm(drug.iterrows(), total=drug.shape[0], desc="Processing drugs"):
    name = row['name']
    idx = row['idx']
    
    # 编码药物名称
    encoded_input = tokenizer(name, return_tensors='pt')

    # 将编码后的输入移到GPU（如果可用）
    encoded_input = {key: value.to(device) for key, value in encoded_input.items()}

    # 获取模型的输出
    with torch.no_grad():  # 不需要计算梯度
        output = model(**encoded_input)

    # 提取[CLS] token的embedding (即output['last_hidden_state'][0][0])
    embedding = output['last_hidden_state'][0][0].cpu().numpy()  # 转回CPU并转换为numpy数组

    # 将生成的embedding保存到字典中，key为id，value为embedding
    embeddings_dict[idx] = embedding

    # print(embedding)

    
# 将字典保存为pickle文件
pickle_path = '/data/zhaojingtong/PrimeKG/data_all/drug_embeddings.pkl'
with open(pickle_path, 'wb') as f:
    pickle.dump(embeddings_dict, f)

print(f"Embeddings saved to {pickle_path}")

```
