# Standardization of molecular orientation

```python
def standardization(data_path,file_name,save_path):
    xyz_file=data_path
    spms = SPMS(xyz_file,key_atom_num=None,desc_n=40,desc_m=40)
    spms._Standarlize_Geomertry()
    new_positions=spms.new_positions
    with open(xyz_file, 'r') as f:
        lines = f.read()
        atom_num=eval(lines.split('\n')[0])
        string=''
        string+=lines.split('\n')[0]
        string+='\n'
        string+=lines.split('\n')[1]
        string+='\n'
        for i in range(2,2+atom_num):
            lines_split=lines.split('\n')[i].split(' ')
            tem_index=[index  for index,i in enumerate(lines_split) if i !='']
            lines_split[tem_index[1]]='%.14f'%new_positions[i-2][0]
            lines_split[tem_index[2]]='%.14f'%new_positions[i-2][1]
            lines_split[tem_index[3]]='%.14f'%new_positions[i-2][2]
            tem_lines = lines_split[tem_index[0]]+add_o_o(lines_split[tem_index[0]],eval(lines.split('\n')[i].split(' ')[tem_index[1]]),new_positions[i-2][0],tem_index[1],tem_index[0])+\
                        lines_split[tem_index[1]]+add_o(eval(lines.split('\n')[i].split(' ')[tem_index[2]]),new_positions[i-2][1],tem_index[2],tem_index[1])+\
                        lines_split[tem_index[2]]+add_o(eval(lines.split('\n')[i].split(' ')[tem_index[3]]),new_positions[i-2][2],tem_index[3],tem_index[2])+\
                        lines_split[tem_index[3]]+'\n'
            string+=tem_lines
        # print(string)
        with open(save_path+file_name,'w') as fw:
            fw.writelines(string)
```

```python
standardization('/project/reproduct_paper/SEMG-MIGNN/standardization/0_xtbopt.xyz','0_xtbopt_std.xyz','/project/reproduct_paper/SEMG-MIGNN/standardization/')
```
